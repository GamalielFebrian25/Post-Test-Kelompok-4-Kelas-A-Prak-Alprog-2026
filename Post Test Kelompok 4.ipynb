```python
import os
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

#Agar kalau ganti device cukup sekali aja ubah path nya
CSV_PATH = r'C:\Users\User\Downloads\Kelas A_Gender Inequality Index.csv'
OUTPUT_DIR = "kelompok4_ipynb_output"

os.makedirs(OUTPUT_DIR, exist_ok=True)
sns.set_theme(style="whitegrid")
```

## Load Dataset

Bagian ini membaca file CSV dan membersihkan kolom numerik agar bisa dihitung dengan benar.

```python
df = pd.read_csv(CSV_PATH)

numeric_columns = [
    "GII",
    "Rank",
    "Maternal_mortality",
    "Adolescent_birth_rate",
    "Seats_parliament",
    "F_secondary_educ",
    "M_secondary_educ",
    "F_Labour_force",
    "M_Labour_force",
]

# Ubah tipe data jadi numerik, ganti isi yang tidak valid dengan NaN (kosong)
for column in numeric_columns:
    df[column] = pd.to_numeric(df[column], errors="coerce")

display(df.head())
```

## Kategori A

Tentukan nilai median dari tingkat kelahiran remaja (`Adolescent_birth_rate`) berdasarkan kelompok `Human_development`. Tampilkan dalam Bar Chart.

```python
# 1. Logika Data
kategori_a = (
    df.groupby("Human_development", as_index=False)["Adolescent_birth_rate"]
    .median()
    .sort_values("Adolescent_birth_rate", ascending=False)
)

kategori_a.to_csv(
    os.path.join(OUTPUT_DIR, "kategori_A_median_adolescent_birth_rate.csv"),
    index=False,
)
display(kategori_a)

# 2. Visualisasi Individu
plt.figure(figsize=(9, 5))
sns.barplot(
    data=kategori_a,
    x="Human_development",
    y="Adolescent_birth_rate",
    palette="viridis",
)
plt.title("Median Adolescent Birth Rate berdasarkan Human Development")
plt.xlabel("Human Development")
plt.ylabel("Median Adolescent Birth Rate")
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "kategori_A_bar_chart.png"), dpi=300)
plt.show()
```

## Kategori B

Cari 10 negara dengan peringkat `Rank` GII terburuk atau tertinggi yang tingkat pendidikan perempuannya (`F_secondary_educ`) masih di bawah 50%. Tampilkan data pendidikannya dalam Bar Chart.

```python
# 1. Logika Data
kategori_b = (
    df[(df["F_secondary_educ"] < 50) & df["Rank"].notna()]
    .sort_values("Rank", ascending=False)
    .head(10)
    [["Country", "Human_development", "Rank", "GII", "F_secondary_educ"]]
)

kategori_b.to_csv(
    os.path.join(OUTPUT_DIR, "kategori_B_rank_terburuk_f_secondary_lt50.csv"),
    index=False,
)
display(kategori_b)

# 2. Visualisasi Individu
plt.figure(figsize=(11, 6))
sns.barplot(
    data=kategori_b,
    x="F_secondary_educ",
    y="Country",
    palette="magma",
)
plt.title("10 Negara dengan Rank GII Terburuk (Pend. Perempuan < 50%)")
plt.xlabel("Pendidikan Menengah Perempuan (%)")
plt.ylabel("Negara")
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "kategori_B_bar_chart.png"), dpi=300)
plt.show()
```

## Kategori C

Lakukan perhitungan matriks korelasi `.corr()` antar seluruh variabel numerik khusus untuk status `Human_development` Medium dan Low. Tampilkan dalam Heatmap.

```python
# 1. Logika Data
df_c = df[df["Human_development"].isin(["Medium", "Low"])]
df_num_c = df_c.select_dtypes(include=[np.number])
korelasi_c = df_num_c.corr()

korelasi_c.to_csv(os.path.join(OUTPUT_DIR, "kategori_C_korelasi_medium_low.csv"))
display(korelasi_c)

# 2. Visualisasi Individu
plt.figure(figsize=(10, 8))
sns.heatmap(
    korelasi_c, 
    annot=True, 
    cmap="coolwarm", 
    fmt=".2f"
)
plt.title("Matriks Korelasi (Medium & Low)")
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "kategori_C_heatmap.png"), dpi=300)
plt.show()
```

## Kategori D

Tampilkan sebaran data kematian ibu (`Maternal_mortality`) antar kelompok `Human_development` dalam Boxplot.

```python
# 1. Logika Data
kategori_d = df.dropna(subset=["Human_development", "Maternal_mortality"])

summary_d = kategori_d.groupby("Human_development")["Maternal_mortality"].describe()
summary_d.to_csv(os.path.join(OUTPUT_DIR, "kategori_D_summary_maternal_mortality.csv"))
display(summary_d)

# 2. Visualisasi Individu
plt.figure(figsize=(9, 6))
sns.boxplot(
    data=kategori_d,
    x="Human_development",
    y="Maternal_mortality",
    palette="Set2",
)
plt.title("Sebaran Maternal Mortality berdasarkan Human Development")
plt.xlabel("Human Development")
plt.ylabel("Maternal Mortality")
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "kategori_D_boxplot.png"), dpi=300)
plt.show()
```

## OUTPUT WAJIB: Grafik 5 (Gabungan)

Menggabungkan seluruh visualisasi Kategori A, B, C, dan D dalam satu layout grid 2x2 menggunakan pendekatan objek `ax=axes[x, y]` tanpa menulis ulang logika datanya (memanggil ulang variabel yang sudah dihitung).

```python
# Membuat layout grid 2x2
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle("Analisis Gender Inequality Index (GII) - Kelompok 4", fontsize=18, fontweight="bold")

# Grafik 1 (Kategori A) di grid kiri atas
sns.barplot(data=kategori_a, x="Human_development", y="Adolescent_birth_rate", ax=axes[0, 0], palette="viridis")
axes[0, 0].set_title("Kategori A: Median Kelahiran Remaja")
axes[0, 0].set_xlabel("Human Development")
axes[0, 0].set_ylabel("Median Adolescent Birth Rate")

# Grafik 2 (Kategori B) di grid kanan atas
sns.barplot(data=kategori_b, x="F_secondary_educ", y="Country", ax=axes[0, 1], palette="magma")
axes[0, 1].set_title("Kategori B: 10 Rank GII Terburuk (Pend. < 50%)")
axes[0, 1].set_xlabel("Pendidikan Menengah Perempuan (%)")
axes[0, 1].set_ylabel("Negara")

# Grafik 3 (Kategori C) di grid kiri bawah
sns.heatmap(korelasi_c, annot=True, cmap="coolwarm", fmt=".2f", ax=axes[1, 0])
axes[1, 0].set_title("Kategori C: Matriks Korelasi (Medium & Low)")

# Grafik 4 (Kategori D) di grid kanan bawah
sns.boxplot(data=kategori_d, x="Human_development", y="Maternal_mortality", ax=axes[1, 1], palette="Set2")
axes[1, 1].set_title("Kategori D: Distribusi Maternal Mortality")
axes[1, 1].set_xlabel("Human Development")
axes[1, 1].set_ylabel("Maternal Mortality")

# Merapikan dan menyimpan grafik gabungan
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "Grafik_5_Gabungan_Kelompok4.png"), dpi=300)
plt.show()
```
